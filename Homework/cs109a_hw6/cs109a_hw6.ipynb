**1.1** In this problem, we will observe how both tree-depth and cross-validation affect our ability to accurately model data. Specifically, for each tree depth from 1 to 20 (inclusive):

- Fit a decision tree to the entire **training** set.

- Evaluate on the entire **training** set (i.e., `.score(...)`), while storing the scores in a variable named `train_scores`.

- Perform 5-fold cross-validation with the entire **training** set, while storing the mean validation score and the validation standard deviation  in variables named `cvmeans` and `cvstds`, respectively.

Now that we have these informative scores, let us plot them. Generate 2 plots, both showing (a) the non-cross-validation training scores, (b) the mean validation scores, and (c) a shaded region that illustrates the +/-2 standard deviation validation bounds for each tree depth. The content and formatting of these 2 plots should be identical, EXCEPT in one plot set the limits on the y-axis to focus on the validation performance. Remember to label and title each plot appropropriately.

**HINT:** You can use `plt.fill_between(...)` to easily generate the shaded region in your plots.

</div>

````python
# your code here 
xtrain = X_train.copy()
ytrain = y_train.copy()
xtest = X_test.copy()

depth = np.linspace(1,20,20)
cvmeans, cvstds, train_scores = [], [], []

folds = 5

for i in depth:
    decisiontree = DecisionTreeClassifier(criterion = "gini", max_depth = i)
    decisiontree.fit(xtrain, ytrain)
    score = decisiontree.score(xtrain,ytrain)
    train_scores.append(score)
    
    # now doing the cross validation
    cs_val = cross_val_score(decisiontree, xtrain, ytrain, cv = folds)
    cvmeans.append(np.mean(cs_val))
    cvstds.append(np.std(cs_val))
````

````python
plt.figure(figsize = (8,8))
plt.plot(depth, train_scores, "y*-" ,label = "Non Cross Validation Train scores")
plt.plot(depth, cvmeans, "bo-" ,label = "Cross Validation Train scores")

#fill plot
plt.fill_between(depth, np.array(train_scores) + 2*np.array(cvstds), np.array(train_scores) - 2*np.array(cvstds),label = "Standard Deviation Non Cross Validation", alpha = 0.2)
plt.fill_between(depth, np.array(cvmeans) + 2*np.array(cvstds), np.array(cvmeans) - 2*np.array(cvstds),label = "Standard Deviation Non Cross Validation", alpha = 0.2)


plt.title("Comparision of the train scores from Cross Validation & non Cross Validation: Not focused on Validation Performance")
plt.xlabel("Max depth of Tree")
plt.xticks(depth)
plt.ylabel('Score of the Model')
plt.legend()
````

````python
plt.figure(figsize = (8,8))
plt.plot(depth, train_scores, "y*-" ,label = "Non Cross Validation Train scores")
plt.plot(depth, cvmeans, "bo-" ,label = "Cross Validation Train scores")

#fill plot
plt.fill_between(depth, np.array(train_scores) + 2*np.array(cvstds), np.array(train_scores) - 2*np.array(cvstds),label = "Standard Deviation Non Cross Validation", alpha = 0.2)
plt.fill_between(depth, np.array(cvmeans) + 2*np.array(cvstds), np.array(cvmeans) - 2*np.array(cvstds),label = "Standard Deviation Non Cross Validation", alpha = 0.2)


plt.title("Comparision of the train scores from Cross Validation & non Cross Validation: Focused on Validation Performance")
plt.ylim((min(cvmeans)-0.01, max(cvmeans)+0.01)) # added/minus 0.01 to better see
plt.xticks(depth)
plt.xlabel("Max depth of Tree")
plt.ylabel('Score of the Model')
plt.legend()
````

````python
# your code here
ytest = y_test.copy()

best_depth = depth[np.argmax(cvmeans)]
decisiontree = DecisionTreeClassifier(criterion = "gini", max_depth = best_depth)
decisiontree.fit(xtrain, ytrain)

ypred_train = decisiontree.predict(xtrain)
ypred_test = decisiontree.predict(xtest)

best_cv_tree_train_score = accuracy_score(ytrain,ypred_train)*100
best_cv_tree_test_score = accuracy_score(ytest, ypred_test)*100

print(f"Here is the train Accuracy score from the depth chosen {best_depth}: {best_cv_tree_train_score:.3f}%")
print(f"Here is the test Accuracy score from the depth chosen {best_depth}: {best_cv_tree_test_score:.3f}%")
````

I chose the best_depth to be 5 based off of the cross validation model. It looks like the max accuracy score was at tree depth of 5 from the graph. This would be the most generalizable to unseen data sets because it isn't fitting to the train set too much, and was validated through cross-validation as a hyper-parameter. Cross validation works to chose this hyper-parameter because it creates multiple folds, in which it fits on certain chunks on the data and validates on the remaining chunk. It repeats this process for each fold to almost test how the model will do on many different actual test sets.

**2.2** Here we will use the `tree_depth` chosen in Question 2.1 to generate 55 boostrapped sets of decision tree predictions for both the training and test data. To accomplish this:

**NOTE:** There is no need to do anything with your `bagging_models_list` list yet. It will not be used until later in [Question 3.2](#part3).

**REQUIRED DATAFRAME STRUCTURE:** The training and test prediction results of your bootstraps should be returned by the `bagger` function as dataframes formatted like the example shown below. Each row should represent one observation (from either the training or test set depending on the dataframe), and each column should represent one bootstrapped result. The values stored in the dataframe are the bootstrapped predictions for each observation as illustrated below.

As an example, the required structure of the `bagging_train_df` and `bagging_test_df` dataframes is:

|     |bootstrap model 1|bootstrap model 2|...|bootstrap model 55|  
| --- | --- | --- | --- | --- |
|0| 0 | 1|... |0|
|1| 1| 1|... |0|
|2| 0| 0|... |1|
|...| ...| ...|... |... |
| $n$-1 | 0| 0|... |1|

**HINT:** You can use `resample(...)` from scikit-learn to easily bootstrap the $X$ and $y$ data.

````python
# Complete this function as instructed
# Do not modify the function signature or docstring
def bagger(
    n_trees: int,
    tree_depth: int,
    X_train: np.ndarray,
    y_train: np.ndarray,
    X_test: np.ndarray,
    random_seed: int = 0,
) -> (pd.DataFrame, pd.DataFrame, list):
    """Generate boostrapped DecisionTreeClassifier predictions
    
    Function fits bootstrapped DecisionTreeClassifier models
    and returns training and test predictions for each of those
    bootstrapped models, along with the fitted model objects as
    described in Question 2.2 question text.
    
    :param n_trees: int, number of bootstrapped decision trees
    :param tree_depth: int, maximum tree depth
    :param X_train: np.ndarray, training X observations
    :param y_train: np.ndarray, training y observations
    :param X_test: np.ndarray, test X observations
    :param random_seed: int, random seed used to set np.random.seed
                        to ensure replicable results (default=0)
    
    :returns: (pd.DataFrame, pd.DataFrame, list), tuple containing 3
              objects, (1) bagging_train_df dataframe
              as described in Q2.2 question text, (2) bagging_test_df
              dataframe as described in Q2.2, and (3) bagging_models_list
              containing every trained DecisionTreeClassifier model
              object (i.e. estimator), one estimator for each bootstrap
              (you will need this list later in Q3.2)
    """
    # your code here
    np.random.seed(random_seed)
    bagging_models_list, bagging_train_list, bagging_test_list = [], [], []

    
    #need to do 55 bootstraps
    for i in range(n_trees):
        # create bootstraps 
        xtrain_new, ytrain_new = resample(X_train,y_train, replace = True) # need a rule
        xtest_new = resample(X_test, replace = True) # need a rule

        #creating model
        model = DecisionTreeClassifier(max_depth = tree_depth)
        model.fit(xtrain_new, ytrain_new) # only fit on bootstraps 
        y_pred_train = model.predict(X_train) # need to predict
        y_pred_test = model.predict(X_test)

        #insert into the dataframe
        bagging_train_list.append(y_pred_train)
        bagging_test_list.append(y_pred_test)

        # appending the model to the list
        bagging_models_list.append(model)

    bagging_train_df = pd.DataFrame(bagging_train_list).T
    bagging_test_df = pd.DataFrame(bagging_test_list).T
 
    
    # end of your code here
    
    return bagging_train_df, bagging_test_df, bagging_models_list
     
````

````python
# RUN THIS CELL
# DO NOT MODIFY THIS CODE

def running_predictions(
    bagger_results_df: pd.DataFrame,
    targets: np.ndarray,
) -> pd.Series:
    """Generates running accuracy of intermediate bootstraps when bagging 
    
    Generates a series of accuracy scores calculated using the
    running predictions of each additional bootstrapped model
    generated using the Question 2.2 `bagger` function. For example,
    the first accuracy in the series represents the predictive
    accuracy of just the first bootstrapped model. The second accuracy
    reflects the bagged accuracy of the first 2 bootstrapped models. The
    j-th accuracy reflects the bagged accuracy of the first j
    bootstrapped models.
    
    :param bagger_results_df: pd.DataFrame, a bagging results dataframe
                              (either train or test) output from the Q2.2
                              `bagger` function
    :param targets: np.ndarray, 1-dimensional array of true class labels
                    for either train or test observations (i.e y_train or
                    y_test, whichever corresponds to the inputted
                    bagger_results_df)             
    :returns: pd.Series, a series of values showing the accuracy of
              using the initial j trees to predict the targets for each
              value of j bootstrapped models
    """
    # verify that input data objects meet the requirements specified
    # in the docstring
    assert type(bagger_results_df)==pd.core.frame.DataFrame, (
        "baggin_results_df input must be a pd.DataFrame"
    )
    assert type(targets)==np.ndarray, (
        "targets input must be an np.ndarray"
    )
    assert targets.ndim==1, (
        "targets input np.ndarray must be one-dimensional"
    )
    
    # identify the number of bootstrapped trees in inputted bagger df
    n_trees = bagger_results_df.shape[1]
    
    # calculate the running percentage of models voting 1 as each
    # additional model is considered
    running_percent_1s = (
        np.cumsum(bagger_results_df, axis=1)/np.arange(1,n_trees+1)
    )
    
    # predict 1 when the running average is above 0.5
    running_conclusions = running_percent_1s > 0.5
    
    # check whether the running predictions match the targets
    running_correctnesss = running_conclusions == targets.reshape(-1,1)
    
    # calculate and return final accuracies
    return np.mean(running_correctnesss, axis=0)
````

````python
# your code here

#not clear what targets are
correct_accuracies_train =  running_predictions(bagging_train_df,y_train)*100
correct_accuracies_test =  running_predictions(bagging_test_df,y_test)*100

plt.figure(figsize = ((10,10)))
plt.plot(np.arange(0,55),correct_accuracies_train, label = "Running Accuracies from Bagging Train Dataframe")
plt.plot(np.arange(0,55),correct_accuracies_test, label = "Running Accuracies from Bagging Test Dataframe")
plt.axhline(y = best_cv_tree_test_score, color = "r", label = "Test Accuracy from Best model in Q1") 
plt.axhline(y = worst_cv_tree_test_score, color = "b", label = "Test Accuracy from worst model chosen in 2.1") 
plt.title("Test and Train accuracies for each Bootstraped Model [1-55]")
plt.xlabel("Number of Bootstraps")
plt.ylabel(" Accuracy Score in %")
plt.legend()
````

````python
# your code here
def boosting(X_train,y_train,X_test, y_test):
    tree = DecisionTreeClassifier(max_depth = 3)
    tree1 = tree.fit(X_train,y_train) # fit on train data 
    y_pred_train = tree1.predict(X_train)
    y_pred_test = tree1.predict(X_train)

    #train data
    matches_train = (y_pred_train == y_train).astype(int)
    amount_correct_train = sum(matches_train)
    accuracy_train = amount_correct_train/len(y_train)

    #test data
    matches_test = (y_pred_test == y_test).astype(int)
    amount_correct_test = sum(matches_test)
    accuracy_test = amount_correct_test/len(y_test)

    print(f"Here is the train accuracy from the first model: {accuracy_train}")
    print(f"Here is the train accuracy from the first model: {accuracy_test}\n")

    new_weights_train = np.where(matches_train==1, 1,2)

    #Second model
    tree2 = tree.fit(X_train, y_train, sample_weight=new_weights_train)

    y_pred2_train = tree2.predict(X_train)
    y_pred2_test = tree2.predict(X_test)

    # calculate for train data
    matches_train2 = (y_pred2_train == y_train).astype(int)
    amount_correct_train2 = sum(matches_train2)
    accuracy_train2 = amount_correct_train2/len(y_train)
    print(f"Here is the test accuracy from the second model: {accuracy_train2}")

    #calculate for test data
    matches_test2 = (y_pred2_test == y_test).astype(int)
    amount_correct_test2 = sum(matches_test2)
    accuracy_test2 = amount_correct_test2/len(y_test)
    print(f"Here is the test accuracy from the second model: {accuracy_test2}")

       
````

````python
# Showing our values from the function
boosting(X_train,y_train,X_test, y_test)
````

**your answer here** \
In boosting, when you have too many trees the trees can become correlated. If we have a bunch a regression models, it will give us an array of response values (y hat) based off inputted predictors. For boosting, we first take a simple model over all data with no weights and then we can calculate the residuals (yhat, which we got from the regression - ytrue) to pass as the weights for the next model.
